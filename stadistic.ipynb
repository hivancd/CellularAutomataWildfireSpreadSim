```python
def test_plot():
    # Create a DataFrame
    df = pd.DataFrame({
        'A': [1, 2, np.nan, 4, 5, 6],
        'B': [1, np.nan, 3, 4, 5, 6],
        'C': [1, 2, 3, 4, 5, 6]
    })

    # Plot a histogram of the DataFrame
    df.hist()
    plt.show()

    # Plot a bar chart of the DataFrame
    df.plot(kind='bar')
    plt.show()

    # Plot a line chart of the DataFrame
    df.plot(kind='line')
    plt.show()

    # Plot a scatter plot of the DataFrame
    df.plot(kind='scatter', x='A', y='B')
    plt.show()

    # Plot a box plot of the DataFrame
    df.plot(kind='box')
    plt.show()

    # Plot a hexbin plot of the DataFrame
    df.plot(kind='hexbin', x='A', y='B')
    plt.show()

    # Plot a pie chart of the DataFrame
    df.plot(kind='pie', y='A')
    plt.show()
    
test_plot()
```

```python
data = np.random.normal(0, 1, 1000)  # Example: 1000 random numbers from a normal distribution
percentile_value = np.percentile(data, 75)
# Sort the data
sorted_data = np.sort(data)

# Calculate the cumulative distribution
cumulative_distribution = np.linspace(0, 1, len(sorted_data))

# Plot the ECDF
plt.plot(sorted_data, cumulative_distribution)

plt.axvline(percentile_value, color='r', linestyle='--', label=f'75th Percentile: {percentile_value:.2f}')

plt.xlabel('Data Values')
plt.ylabel('Cumulative Distribution')
plt.title('Empirical Cumulative Distribution Function')
plt.legend()

plt.show()
```
