```python
import matplotlib.pyplot as plt
from perlin_noise import PerlinNoise
import numpy as np

xpix, ypix = 200, 200

noise = PerlinNoise(octaves=7, seed=np.random.randint(1))
pic = [[noise([i/xpix, j/ypix]) for j in range(xpix)] for i in range(ypix)]
noise2 = PerlinNoise(octaves=3, seed=np.random.randint(1))
pic2 = [[noise2([i/xpix, j/ypix]) for j in range(xpix)] for i in range(ypix)]

# noise2 = PerlinNoise(octaves=2, seed=np.random.randint(1))
# pic2 = [[noise2([i/xpix, j/ypix]) for j in range(xpix)] for i in range(ypix)]

#pic = np.array(pic) > abs(np.array(pic2))
lakes = (np.array(pic) > np.ones((xpix, ypix))*0.3) & (np.array(pic) < np.ones((xpix, ypix))*0.7)
b = (np.array(pic2) > np.ones((xpix, ypix))*0) & (np.array(pic2) < np.ones((xpix, ypix))*0.03)
c = (np.array(pic) > np.ones((xpix, ypix))*-0.3) & (np.array(pic) < np.ones((xpix, ypix))*0)
d = (np.array(pic) > np.ones((xpix, ypix))*-1) & (np.array(pic) < np.ones((xpix, ypix))*-0.5)
e = np.array(pic) > np.array(pic2)

plt.imshow(lakes)
plt.show()
plt.imshow(b, cmap='gray')
plt.show()
plt.imshow(c, cmap='gray')
plt.show()
plt.imshow(d, cmap='gray')
plt.show()
plt.imshow(e, cmap='gray')
plt.show()
plt.imshow(pic)
plt.show()
plt.imshow(pic2)
plt.show()
```
